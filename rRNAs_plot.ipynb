```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read custom correct data
rrna = pd.read_excel('input/rrna.xlsx')
rrna.head(2)
```

### Make a plot

```python
plt.figure(figsize=(7, 5))

plt.scatter(rrna['bacteria'], rrna['ref_value'], color='green', marker='s', s=15, label='rRNA in reference')
plt.scatter(rrna['bacteria'], rrna['5S'], color='blue', s=5,label='5S rRNA')
plt.scatter(rrna['bacteria'], rrna['16S'], color='blue', s=5,label='16S rRNA')
plt.scatter(rrna['bacteria'], rrna['23S'], color='blue', s=5,label='23S rRNA')

plt.xlabel('Bacteria')
plt.ylabel('Number of rRNA genes')
plt.title('Number of rRNA genes in reference genome and at the assembly'+"\n")
plt.xticks(rotation=90)
plt.legend()

plt.savefig('results/rrna.svg', bbox_inches='tight')
plt.savefig('results/rrna.png', dpi=300, bbox_inches='tight')
```
